```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import datetime as dt
import zipfile
import pandas as pd
```

```python
zf = zipfile.ZipFile('archive (2).zip') # reading of zip file
# if you want to see all files inside zip folder
zf.namelist() 
```

```python
# now read your csv file 
df = pd.read_csv(zf.open('covid_19_india.csv'))
df.tail(10)
```

```python
Today=df[df.Date=='2021-08-11']
Today
```

```python
#sorting data with respect to max confirmed cases
max_confirmed_cases=Today.sort_values(by='Confirmed',ascending=False)
max_confirmed_cases
```

```python
#sorting data with respect to max death cases
max_death_cases=Today.sort_values(by='Deaths',ascending=False)
max_death_cases
```

```python
top_ten_state=max_confirmed_cases[0:10]
top_ten_state
```

```python
#making bar_chart with most confirmed cases
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x='State/UnionTerritory',y='Confirmed',data=top_ten_state,hue='State/UnionTerritory')
plt.show()
```

```python
top_five_state=max_death_cases[0:7]
top_five_state
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x='State/UnionTerritory',y='Confirmed',data=top_five_state,hue='State/UnionTerritory')
plt.show()
```

```python
max_Cured_state=Today.sort_values(by='Cured',ascending=False)
max_Cured_state
```

```python
max_cured_state=max_Cured_state[0:5]
max_cured_state
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x='State/UnionTerritory',y='Cured',data=max_cured_state,hue='State/UnionTerritory')
plt.show()
```
